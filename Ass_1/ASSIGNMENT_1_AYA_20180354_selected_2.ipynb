#**Normal Distribution**



$P(x;\mu,\sigma)=\displaystyle \frac{1}{\sqrt{2 \pi \sigma^2}} \exp{\displaystyle \left( -\frac{(x-\mu)^2}{2 \sigma^2} \right) }
\hspace{1in}$ 

```python
def normal_pdf(x, mu=0.0, sigma=1.0):
    sqrt_two_pi = math.sqrt(2 * math.pi)
    e= (math.exp(-(x-mu) ** 2 / 2 / sigma ** 2) / (sqrt_two_pi * sigma))
    return e
```

## **TASK**

```python
output=normal_pdf(.5, mu=0.0, sigma=1.0)
output
```

# **Uniform Distribution**

$$p(x) = \begin{cases} \frac{1}{b-a} & x \in [a, b], \\ 0 & x \not\in [a, b].\end{cases}$$

```python
def uniform_pdf(x,a,b):
  if x >= a and x <= b:
    return 1.0/(b-a) 
  else:
    return 0
```

# **Binomial Distribution**


$$
P(r;N,p)=\displaystyle \left(
\frac{N!}{r!(N-r)!}
\right)
p^r (1-p)^{N-r}
$$

#### mean=$Np$, variance=$Np(1-p)$

```python
n=50
p=.6

def factorial(n):
    x = 1
    for i in range(1, n+1):
        x *= i
    return x

def combination(n, r):
    return factorial(n)/(factorial(r)*factorial(n-r))

def binomial(r,n,p):
    return combination(n,r)*(p**r)*((1-p)**(n-r))

rang = list(range(n + 1))
dist = [binomial(r, n, p) for r in rang]

```

# **Beta Distribution**


$$
P(x;α,β)=\displaystyle \left(
const
\right)
*x^{α-1} *(1-x)^{β-1}
$$ 

$$
const = Γ(α + β) / [Γ(α) * Γ(β)]
$$

```python
import math
def const(alpha, beta):
  return math.gamma(alpha) * math.gamma(beta) / math.gamma(alpha + beta)
def beta_pdf(x, alpha, beta):
  if x < 0 or x > 1: 
    return 0
  return (x ** (alpha - 1) * (1 - x) ** (beta - 1) / const(alpha, beta))
```

# Central limit theorm on Normal distribution(m=3,v=2)

```python
f = plt.figure(figsize=(20, 12))
 
def plotHist(nr, N, n_, mean, var0, x0):
    ''' plots the RVs'''
    x = np.zeros((N))
    sp = f.add_subplot(2, 2, n_ )
    
    for i in range(N):    
        for j in range(nr):
            x[i] += np.random.randn() 
        x[i] *= 1/nr
    plt.hist(x, 100,density=True,color='#348ABD', label=" %d RVs"%(nr),alpha=.8);
    plt.setp(sp.get_yticklabels(), visible=False)
    
    variance = var0/nr                     
    fac = 1/np.sqrt(2*np.pi*variance)
    dist = fac*np.exp(-(x0-mean)**2/(2*variance))
    plt.plot(x0,dist,color='#A60628',linewidth=3,label='CLT')
    plt.xlabel('r')
    plt.xlim([-10, 10])
    leg = plt.legend(loc="upper right")
    leg.get_frame().set_alpha(.1)

    
N = 5000  # number of samples taken 
nr = ([1, 5, 10, 25])

mean, var0 =3,2
x0 = np.linspace(-10, 10, 128)
for i in range(np.size(nr)):
    plotHist(nr[i], N, i+1, mean, var0, x0)

plt.suptitle("Addition of independent Gaussian distributions is a Gaussian distribution",fontsize=20);
```

# Central limit theorm on uniform distribution(0,2)

```python
import random
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

    
```

```python
def randomAvg(n=2):
    '''This function draws n random numbers between 0 and 2,
    and returns their average.'''
    avg = 0
    for i in range(n):
        avg += random.uniform(0,2)
    return avg/n
```

```python
myDict = {}
for num in [1,5,10,25]:
    myAvgs = []
    for j in range(5000):
        avg = randomAvg(num)
        myAvgs.append(avg)
        
    
    myDict[num] = myAvgs
```

```python
plt.figure(figsize=(20,12))
sns.set(style="darkgrid")
plt.hist(myDict[1], bins=200,color='indigo')
plt.hist(myDict[5], bins=200,color='darkgreen')
plt.hist(myDict[10], bins=200, alpha=0.7,color='navy')
plt.hist(myDict[25],bins=200, alpha=0.5,color='deeppink')
plt.xlim([0,2])
plt.text(0.31, 31, 'N = 1', fontsize=17)
plt.text(1.3, 49, 'N = 5', fontsize=17)
plt.text(1.15, 65, 'N = 10', fontsize=17)
plt.text(0.99, 89, 'N = 25', fontsize=17)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Value', fontsize=18)
plt.ylabel('Count', fontsize=18)
plt.text(0.05, 89, 'A demo of Central Limit Theorem', fontsize=20)
plt.show()
```

```python
import numpy
import matplotlib.pyplot as plt
f = plt.figure(figsize=(18, 10))
 
def plotHist(nr, N, n_, mean, var0, x0):
    ''' plots the RVs'''
    x = np.zeros((N))
    sp = f.add_subplot(2, 2, n_ )
    
    for i in range(N):    
        for j in range(nr):
            x[i] += np.random.random() 
        x[i] *= 1/nr
    plt.hist(2*x,16,density=True,  color='#348ABD', label=" %d RVs"%(nr))
    plt.setp(sp.get_yticklabels(), visible=False)
    
    variance = var0/nr                     

    fac = 1/np.sqrt(2*np.pi*variance)
    dist = fac*np.exp(-(x0-mean)**2/(2*variance))
    plt.plot(2*x0,.5*dist,color='red',linewidth=3,label='CLT',alpha=.8)
    plt.xlabel('range')
    plt.xlim([0,2])
    leg = plt.legend(loc="upper left")
    leg.get_frame().set_alpha(0.1)
    


N = 5000  # number of samples taken
nr = ([1,5,10,25])
 # mean and variance of uniform distribution in range 0, 2

mean, var0 = .5, 1.0/12

x0 = np.linspace(-2, 2, 128)

for i in range(np.size(nr)):
    plotHist(nr[i], N, i+1, mean, var0, x0)
    

plt.suptitle("Addition of uniform random variables (RVs) converge to a Gaussian distribution (CLT)",fontsize=20);
```

# Central limit on Beta Distribution(0.7,.7)

```python
import scipy.stats as sps
import numpy as np
import matplotlib.pyplot as plt

def sum_random_variables(*kwarg, sp_distribution, n):
  

    v = [sp_distribution.rvs(*kwarg, size=5000) for _ in range(n)]
    return np.sum(v, axis=0)
```

```python
plt.figure(figsize=(20, 8))
N = [1,5,10,25]
for i in range(len(N)):
    plt.subplot(2, 2,i+1) 
    
    s = sum_random_variables(.7, .7, sp_distribution=sps.beta, n=N[i])
  
    plt.hist(s, bins=100,density=True,label=str(N[i]),alpha=.7,color='indigo')
    plt.legend()
plt.suptitle("Beta Distribution with n=1,n=5,n=10,n=25 , N=5000 ",fontsize=20);
plt.show()
```
